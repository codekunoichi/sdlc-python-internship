Solution to the problem above posed as MOTIVATION

```python
def max_n(x):
    
    maximum = x[0]
    for i in range(1, len(x)):
        if x[i] > maximum:
            maximum = x[i]
    
    return(maximum)
```

# TUPLES
Data structure similar to list but with reduced functionality

```python
# Always use parenthesis to denote start and end of tuple
zoo = ('python', 'elephant', 'penguin')
```

```python
# Use len to determine number of elements in tuple
print('Number of animals in the zoo is', len(zoo))
```

```python
# A tuple is NOT MUTABLE like a list
# New tuple has to be created to add elements to existing one
new_zoo = 'monkey', 'camel', zoo
```

```python
# Determine the number of elements of the new tuple
# Note: This is NOT 5
print('Number of cages in the new zoo is', len(new_zoo))
```

```python
# Reference ALL the elements of the previous tuple zoo inside the new tuple new_zoo
print('Animals brought from old zoo are', new_zoo[2])
```

```python
# Reference the last element of the previous tuple zoo inside the new tuple new_zoo
print('Last animal brought from old zoo is', new_zoo[2][2])
```

```python
# Determine the TOTAL number of elements of both tuples
print('Number of animals in the new zoo is',
      len(new_zoo)-1+len(new_zoo[2]))
```

TUPLES ARE MORE MEMORY EFFICIENT THAN LISTS. 
IF YOU DO NOT NEED TO CHANGE YOUR DATA THEN DECLARE IT AS A TUPLE

# DICTIONARIES
Similar to address book where you can find address or contact details of person given her/his name

Associates KEYS (e.g. name) with VALUES (e.g. details like address, phone number and other contact details)

```python
# Specify a dictionary

ab = {
    'Swaroop': 'swaroop@swaroopch.com',
    'Larry': 'larry@wall.org',
    'Matsumoto': 'matz@ruby-lang.org',
    'Spammer': 'spammer@hotmail.com'
}
```

```python
type(ab)
```

```python
# print this dictionary
print(ab)
```

```python
# Determine the number of elements in the dictionary
print(len(ab))
```

```python
# Extract a VALUE given the KEY
print("Swaroop's address is", ab['Swaroop'])
```

```python
# Deleting a KEY-VALUE pair
del ab['Spammer']
```

```python
# Determine the number of elements remaining in the dictionary
print(len(ab))
```

```python
# Adding a new entry i.e. a KEY-VALUE pair
ab['Guido'] = 'guido@python.org'
```

```python
# check what the updated dictionary contains
print(ab)
```
