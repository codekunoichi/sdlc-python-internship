# Control flow with if and while

# Note the indentation

```python
number = 23

# The "input" function in python allows user input
# The "int" function is to convert the user provided input to integer
guess = int(input('Enter an integer : '))

if guess == number:
    # New block starts here
    print('Congratulations, you guessed it.')
    print('(but you do not win any prizes!)')
    # New block ends here
elif guess < number:
    # Another block
    print('No, it is a little higher than that')
    # You can do whatever you want in a block ...
else:
    print('No, it is a little lower than that')
    # you must have guessed > number to reach here

print('Done')
# This last statement is always executed,
# after the if statement is executed.
```

```python
number = 23
running = True

while running:
    guess = int(input('Enter an integer : '))

    if guess == number:
        print('Congratulations, you guessed it.')
        # break
        # this causes the while loop to stop
        running = False
    elif guess < number:
        print('No, it is a little higher than that.')
    else:
        print('No, it is a little lower than that.')
else:
    print('The while loop is over.')
    # Do anything else you want to do here

print('Done')
```

```python
# while loop without else

number = 23
running = True

while running:
    guess = int(input('Enter an integer : '))

    if guess == number:
        print('Congratulations, you guessed it.')
        # this causes the while loop to stop
        running = False
    elif guess < number:
        print('No, it is a little higher than that.')
    else:
        print('No, it is a little lower than that.')

print('Done')
```

```python
# Loop control and termination
# Case 2: "continue" to next iteration of loop
```

```python
while True:
    s = input('Enter something : ')
    if s == 'quit':
        break
    if len(s) < 3:
        print('Too small')
        continue
    print('Length of the string is', len(s))
    print('Input is of sufficient length')
```

```python
# You can define a function to accept user inputs
# a and b are the arguments of the function

def print_max(a, b):
    if a > b:
        print(a, 'is maximum')
    elif a == b:
        print(a, 'is equal to', b)
    else:
        print(b, 'is maximum')
```

```python
# Case 1: We pass constants as arguments to the function print_max

print_max(3, 4)
```

```python
# Case 2: We pass variables as arguments to the function print_max

x = 5
y = 7

print_max(x, y)
```

```python
# Be careful of how variables are defined INSIDE and OUTSIDE functions

# Here x is declared OUTSIDE the function called func and then a variable with the same name x is set INSIDE 
# the function

# However once we come outside the function the value of x set INSIDE the function is no longer valid
# You get back the value set OUTSIDE the function

x = 50


def func(x):
    print('x is', x)
    x = 2
    print('Changed local x to', x)
```

```python
print(x)
func(x)
print('After return from function x is still', x)
```

```python
# Here x is declared OUTSIDE the function called func and then a variable with the same name x is set INSIDE 
# the function but this time it is set as global

# NOTICE THE CHANGE IN BEHAVIOR IN THIS CASE

# Once we come outside the function the value of x OUTSIDE the function stays the same as that what was set 
# INSIDE the function 

x = 50


def func():
    global x

    print('x is', x)
    x = 2
    print('Changed global x to', x)

```

```python
print(x)
func()
print('After return from function value of x  is', x)
```
